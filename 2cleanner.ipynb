```python
def oldDate(n):
    return datetime.now() - timedelta(n)

def path(i, j):
    if j == 0:
        return f'P:/DEPROJECT/Raw_bhav_copy/Raw_bhavcopy_{oldDate(i).strftime("%Y-%m-%d")}.csv'
    else:
        return f'P:/DEPROJECT/Clean_bhav_copy/Clean_bhavcopy_{oldDate(i).strftime("%Y-%m-%d")}.csv'
```

```python
for x in range(1,40):
    try:
        raw_df = pd.read_csv(path(x,0))

        #Reanming columns
        old_col_list = [' SYMBOL',' SERIES',' DATE1',' PREV_CLOSE',' OPEN_PRICE',' HIGH_PRICE',
                        ' LOW_PRICE',' LAST_PRICE',' CLOSE_PRICE',' AVG_PRICE',' TTL_TRD_QNTY',
                        ' TURNOVER_LACS',' NO_OF_TRADES',' DELIV_QTY',' DELIV_PER']

        for j in old_col_list: 
            raw_df.rename(columns = {j:j.strip()}, inplace = True)


        #Deleting 'Turnover_Lacs' column which is not needed 
        raw_df.drop('TURNOVER_LACS', axis=1, inplace=True) 
        raw_df.rename(columns={'DATE1':'DATE'}, inplace=True)


        # Keeping companies who comes under equity section
        raw_df = raw_df.loc[raw_df['SERIES'] == ' EQ'] 
        raw_df.reset_index(drop=True, inplace=True)
        
        # Creating temp datafrmae this data frame will keep records of company which are in our list 
        temp_df = pd.DataFrame()
        for z in Company_list:
            temp_df = temp_df.append(raw_df.loc[raw_df['SYMBOL'] == z], ignore_index=True) 
        print(f'{oldDate(x).strftime("%Y-%m-%d")} cleaned succesfully')
        # Saving clean temp_df to Local system in Clean_bhovcopy file 
        temp_df.to_csv(path(x,1),index=False)
    except:
        pass
```
