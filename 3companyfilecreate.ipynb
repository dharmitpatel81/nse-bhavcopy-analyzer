```python
def oldDate(n):
    return datetime.now() - timedelta(n)
def path(i):
        return f'P:/DEPROJECT/Clean_bhav_copy/Clean_bhavcopy_{oldDate(i).strftime("%Y-%m-%d")}.csv'
def Comapny_file_path(name):
    return f'P:/DEPROJECT/Company_file/{name}.csv'
```

```python
for i in Company_list:
    df = pd.DataFrame(columns=COLUMN_NAMES)
    df.to_csv(f'P:/DEPROJECT/Company_file/{i}.csv',index=False)
```

```python
for i in reversed(range(0,40)):
    try:
        df = pd.read_csv(path(i))
        print(oldDate(i).strftime("%Y-%m-%d"))
        for i in Company_list:
            # Loading companys csv file in program 
            company_df = pd.read_csv(Comapny_file_path(i))
            
            # Finding comapnys row from clean bhav copy 
            temp_df = df[df['SYMBOL']== i]

            # Adding companys now to company's csv file 
            company_df = pd.concat([company_df, temp_df], ignore_index = True, axis = 0)

            company_df.drop_duplicates(subset ="DATE",keep = "last", inplace = True,ignore_index= True)    
            company_df.to_csv(Comapny_file_path(i), index=False)
    except:
        print(f'{oldDate(i).strftime("%Y-%m-%d")} --> no data')
```
