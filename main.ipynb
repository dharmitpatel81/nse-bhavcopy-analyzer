```python
def url(i) :
    return f'https://archives.nseindia.com/products/content/sec_bhavdata_full_{oldDate(i).strftime("%d%m%Y")}.csv'

def oldDate(n):
    return datetime.now() - timedelta(n)

def path(i,j):
    if j == 0:
        return f'Raw_bhav_copy/Raw_bhavcopy_{oldDate(i).strftime("%Y-%m-%d")}.csv'
    else:
        return f'Clean_bhav_copy/Clean_bhavcopy_{oldDate(i).strftime("%Y-%m-%d")}.csv'
    
def Comapny_file_path(name):
    return f'Company_file/{name}.csv'

def Analyzed_company_file_path(name):
    return f'Analyzed_Company_file/{name}.csv'
```

```python
def rawdown(x):
    for i in (range(0,x)):
        try: 
            r = requests.get(url(i), timeout=1)
            print(f"Bhavcopy of {oldDate(i).strftime('%Y-%m-%d')} downloaded")
            df = pd.read_csv(url(i))
            df.to_csv(path(i,0),index=False)
        except requests.exceptions.Timeout as err: 
            pass
        
x = int(input("Enter Number of days back you want to perform operation: "))
rawdown(x)
```

# CLEANING

```python
for i in range(1,x):
    try:
        raw_df = pd.read_csv(path(i,0))

        #Reanming columns
        old_col_list = [' SYMBOL',' SERIES',' DATE1',' PREV_CLOSE',' OPEN_PRICE',' HIGH_PRICE',
                        ' LOW_PRICE',' LAST_PRICE',' CLOSE_PRICE',' AVG_PRICE',' TTL_TRD_QNTY',
                        ' TURNOVER_LACS',' NO_OF_TRADES',' DELIV_QTY',' DELIV_PER']

        for j in old_col_list: 
            raw_df.rename(columns = {j:j.strip()}, inplace = True)


        #Deleting 'Turnover_Lacs' column which is not needed 
        raw_df.drop('TURNOVER_LACS', axis=1, inplace=True) 
        raw_df.rename(columns={'DATE1':'DATE'}, inplace=True)


        # Keeping companies who comes under equity section
        raw_df = raw_df.loc[raw_df['SERIES'] == ' EQ'] 
        raw_df.reset_index(drop=True, inplace=True)
        
        # Creating temp datafrmae this data frame will keep records of company which are in our list 
        temp_df = pd.DataFrame()
        for z in Company_list:
            temp_df = temp_df.append(raw_df.loc[raw_df['SYMBOL'] == z], ignore_index=True) 
        print(f'{oldDate(i).strftime("%Y-%m-%d")} cleaned succesfully')
        # Saving clean temp_df to Local system in Clean_bhovcopy file 
        temp_df.to_csv(path(i,1),index=False)
    except:
        pass
```

# Analyzing company file starts from here

```python
def analyzer(df):

    df["DELIV_VALUE"] = ''
    df["5DDeVA"] = ''
    df['%Change_In_Price'] = ''
    df['%Change_In_Deliv_vs_AVG'] = ''
    
    df["DELIV_VALUE"]= (df['DELIV_QTY'] * df['AVG_PRICE']) / 10000000

#     for i in range (5):
#         df['5DDeVA'][i] = np.nan
#         df['%Change_In_Price'][i] = np.nan
#         df['%Change_In_Deliv_vs_AVG'][i] = np.nan
#         df['Intraday_volume'] = np.nan

    for i in range (5, df.shape[0]):
        df.at[i,'5DDeVA'] = df["DELIV_VALUE"].iloc[i-5:i].mean()
        df.at[i,'%Change_In_Price'] = ((df.at[i,'CLOSE_PRICE']-df.at[i-1,'CLOSE_PRICE'])/df.at[i-1,'CLOSE_PRICE']) * 100
        df.at[i,'%Change_In_Deliv_vs_AVG'] = (df.at[i, 'DELIV_VALUE']/df.at[i, '5DDeVA']) * 100 
        df.at[i,'Intraday_volume'] = int(df.at[i,'TTL_TRD_QNTY']-df.at[i,"DELIV_QTY"])

    return df
```

```python
for company in Company_list:

    df = pd.read_csv(Comapny_file_path(company))

    analyzer(df).to_csv(Analyzed_company_file_path(company), index=False)
    print(f'{company} file analyzed')
```
